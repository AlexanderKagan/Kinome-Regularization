```python
v0_list = [1e-06, 1e-05, 1e-04, 0.001, 0.003, 0.1, 0.3, 1, 3, 10, 30, 60, 80, 100]
# v0_list = [0.001]
```

```python
y_test_predicted_list = []
y_test_true_standartized_list = []
for i in v0_list:
    if i <= 1e-04:
        i = format(i, '.0e')
    print(i)
    folder_path = f'./fof/data/r/validation_output/plots/{i}/'
    os.makedirs(folder_path, exist_ok=True)
    
    y_train_predicted = pd.read_csv(f'./fof/data/r/validation_output/y_train_predicted_{i}.csv')['x'].to_numpy()
    y_train_true_standartized = pd.read_csv(f'./fof/data/r/validation_output/y_train_true_standartized_{i}.csv')['x'].to_numpy()
    fig = px.scatter(x=y_train_predicted, y=y_train_true_standartized, title=f'v0: {i}, Train')
    fig.update_xaxes(title_text="Predicted Values")  # X-axis label
    fig.update_yaxes(title_text="True Standarized Values")  # Y-axis label
    fig.show()
    fig.write_html(folder_path + 'true_predicted_scatter_plot_train.html')
    
    y_test_predicted = pd.read_csv(f'./fof/data/r/validation_output/y_test_predicted_{i}.csv')['x'].to_numpy()
    y_test_predicted_list.append(y_test_predicted)
    y_test_true_standartized = pd.read_csv(f'./fof/data/r/validation_output/y_test_true_standartized_{i}.csv')['x'].to_numpy()
    y_test_true_standartized_list.append(y_test_true_standartized)
    fig = px.scatter(x=y_test_predicted, y=y_test_true_standartized, title=f'v0: {i}, Test')
    fig.update_xaxes(title_text="Predicted Values")  # X-axis label
    fig.update_yaxes(title_text="True Standarized Values")  # Y-axis label
    fig.show()
    fig.write_html(folder_path + 'true_predicted_scatter_plot_test.html')
```

```python
mse_list = []
for i, (y_predicted, y_true) in enumerate(zip(y_test_predicted_list, y_test_true_standartized_list)):
    mse_list.append(mean_squared_error(y_predicted, y_true))
```

```python
px.scatter(x=[str(i) for i in v0_list], y=mse_list)
```

```python
fig = px.imshow(difference_matrix, x=[str(i) for i in v0_list],
               y=[f'{i} - {v}' for i, v in enumerate(v0_list)],
               color_continuous_scale='Viridis')  # You can choose a different color scale

# Add labels to the axes
fig.update_xaxes(side="top")  # To display the x-axis labels at the top
fig.update_layout(xaxis_title="Columns", yaxis_title="Rows")
```
